## Métricas de evaluación en regresión lineal simple



Para evaluar la calidad de un modelo de regresión lineal simple, se utilizan varias métricas:

- **$R^2$ (Coeficiente de determinación):**  
  Indica la proporción de la variabilidad de la variable dependiente $Y$ explicada por la variable independiente $X$.  
  $$
  R^2 = 1 - \frac{\text{Suma de residuos al cuadrado}}{\text{Suma total de cuadrados}}
  $$
  Un valor cercano a 1 indica un mejor ajuste.

- **MSE (Error cuadrático medio):**  
  Promedio de los errores al cuadrado entre los valores predichos y los reales.
  $$
  MSE = \frac{1}{n} \sum (y_{i} - \hat{y}_{i})^{2}
  $$

- **MAE (Error absoluto medio):**  
  Promedio del valor absoluto de los errores.
  $$
  MAE = \frac{1}{n} \sum |y_{i} - \hat{y}_{i}|
  $$

- **AIC y BIC:**  
  Métricas que ayudan a comparar modelos, considerando el equilibrio entre ajuste y complejidad.

Estas métricas permiten comparar modelos y elegir el que mejor se ajusta a los datos y necesidades del análisis.

## Medidas de incertidumbre: Intervalos de confianza y valores p



Los coeficientes estimados en un modelo de regresión tienen asociada una incertidumbre, que se mide mediante:

- **Intervalos de confianza:**  
  Un intervalo de confianza del 95% significa que, si repitiéramos el muestreo y el ajuste del modelo muchas veces bajo las mismas condiciones, aproximadamente el 95% de los intervalos calculados contendrían el valor real del parámetro poblacional.  
  Es decir, el intervalo no indica la probabilidad de que el parámetro esté dentro de ese rango para una muestra específica, sino la proporción de intervalos que capturarían el valor verdadero si el procedimiento se repite muchas veces.

- **Valor p:**  
  Probabilidad de obtener un resultado tan extremo como el observado si la hipótesis nula es verdadera. Un valor p bajo (menor a 0.05) indica que el coeficiente es estadísticamente significativo.

Interpretar correctamente estas medidas es fundamental para tomar decisiones informadas basadas en el modelo.

## Intervalos de confianza en el resumen y personalizados



El método `summary()` de statsmodels siempre muestra los intervalos de confianza al 95%. Si necesitas ver los intervalos de confianza al 99% (u otro nivel), debes calcularlos aparte con `model.conf_int(alpha=...)`.
A continuación se muestran ambos resultados para comparar: el resumen estándar (95%) y los intervalos de confianza al 99%.

`````python
# Mostrar el resumen estándar (IC 95%)
print(model.summary())

# Mostrar los intervalos de confianza al 99%
print("\nIntervalos de confianza al 99%:")
print(model.conf_int(alpha=0.01))
`````

### Intervalos de confianza del 99% para las predicciones
Puedes obtener intervalos de confianza al 99% para las predicciones del modelo usando el método `get_prediction()` y el método `summary_frame(alpha=0.01)`. Esto te permite ver el rango de incertidumbre para cada predicción individual.

`````python
# Obtener predicciones y sus intervalos de confianza al 99% para cada observación
pred = model.get_prediction(data)
pred_summary = pred.summary_frame(alpha=0.01)  # 99% de confianza
pred_summary.head()  # Mostramos solo las primeras filas
`````

**Resumen del modelo OLS con intervalos de confianza al 99% para los coeficientes**

La tabla siguiente muestra los coeficientes, errores estándar, valores t, p y los intervalos de confianza al 99% para cada parámetro del modelo. Además, se muestran las métricas principales del ajuste (R², R² ajustado, F, Prob(F), etc.).

`````python
import pandas as pd

# Extraer los valores principales del modelo
summary_dict = {
    'Coef.': model.params,
    'Std.Err.': model.bse,
    't': model.tvalues,
    'P>|t|': model.pvalues,
}
# Intervalos de confianza al 99%
conf_int_99 = model.conf_int(alpha=0.01)
summary_df = pd.DataFrame(summary_dict)
summary_df[['[0.005', '0.995]']] = conf_int_99

# Mostrar tabla
print('Métricas principales:')
print(f'R²: {model.rsquared:.4f}')
print(f'R² ajustado: {model.rsquared_adj:.4f}')
print(f'F-statistic: {model.fvalue:.2f}')
print(f'Prob (F-statistic): {model.f_pvalue:.4g}')
print(f'Log-Likelihood: {model.llf:.2f}')
print(f'AIC: {model.aic:.2f}')
print(f'BIC: {model.bic:.2f}\n')

print('Tabla de coeficientes con IC al 99%:')
display(summary_df)
`````
