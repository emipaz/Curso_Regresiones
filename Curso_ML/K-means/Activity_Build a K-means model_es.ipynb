### Escalar las características

Porque K-means usa la distancia entre observaciones como su medida de similitud, es importante escalar los datos antes de modelar. Use una herramienta de terceros, como la función `StandardScaler` de scikit-learn. `StandardScaler` escala cada punto xᵢ restando el valor medio observado para esa característica y dividiendo por la desviación estándar:

x-escalado = (xᵢ – media(X)) / σ

Esto asegura que todas las variables tengan una media de 0 y una varianza/desviación estándar de 1.
