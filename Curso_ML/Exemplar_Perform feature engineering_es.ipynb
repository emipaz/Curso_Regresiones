- Deberías evitar seleccionar la columna `name` como característica. El nombre de un jugador no ayuda a determinar la duración de su carrera. Además, puede no ser ético o justo predecir la duración de la carrera de un jugador basándose en un nombre.
- El número de partidos en los que un jugador ha participado puede no ser tan importante para determinar la duración de su carrera como la cantidad de puntos que ha obtenido. Aunque podrías decir que alguien que ha jugado en más partidos puede tener más práctica y experiencia, los puntos que obtiene durante los partidos en los que jugó hablarían más de su rendimiento como jugador. Esto, a su vez, influiría en la duración de su carrera. Entonces, la columna `gp` por sí sola puede no ser una característica útil. Sin embargo, `gp` y `pts` podrían combinarse para obtener el total de puntos obtenidos en todos los partidos, y ese resultado podría ser una característica útil. Ese enfoque puede implementarse más adelante en el proceso de ingeniería de características—en la extracción de características.
- Si se extraerá como característica el número de puntos obtenidos en los partidos, entonces eso podría combinarse con el número de minutos jugados en los partidos (`min * gp`) para extraer otra característica. Esto podría ser una medida de eficiencia de los jugadores y podría ayudar a predecir la duración de la carrera de los jugadores. `min` por sí solo puede no ser útil como característica por la misma razón que `gp`.

Esto permite una comparación más significativa entre jugadores. La misma lógica se aplica al porcentaje de tiros de tres puntos encestados, así como al porcentaje de tiros libres encestados.
- Hay columnas para el número de rebotes ofensivos (`oreb`), el número de rebotes defensivos (`dreb`) y el número total de rebotes (`reb`). Debido a que el número total de rebotes ya debería incorporar tanto los rebotes ofensivos como los defensivos, tendría sentido usar el total como una característica.

Pregunta: ¿Qué columnas se prestan para la extracción de características?

- Las columnas `gp`, `pts`, `min` se prestan para la extracción de características.
  - `gp` representa el número total de partidos que un jugador ha jugado, y `pts` representa el promedio de puntos que el jugador ha obtenido por partido. Podría ser útil combinar estas columnas para obtener el total de puntos que el jugador ha obtenido en los partidos y usar el resultado como una nueva característica, que podría añadirse en una nueva columna llamada `total_points`. Los puntos totales obtenidos por un jugador pueden reflejar su rendimiento y moldear la longevidad de su carrera.
  - La columna `min` representa el promedio de minutos jugados por partido. `total_points` podría combinarse con `min` y `gp` para extraer una nueva característica: puntos obtenidos por minuto. Esto puede considerarse una medida de eficiencia del jugador, que podría moldear la duración de la carrera. Esta característica puede añadirse en una columna llamada `efficiency`.


Extrae dos características que crees que ayudarían a predecir `target_5yrs`. Luego, crea una nueva variable llamada 'extracted_data' que contenga las características de 'selected_data', así como las características que se están extrayendo.


```python
# Extract two features that would help predict target_5yrs.
# Create a new variable named `extracted_data`.

### YOUR CODE HERE ### 

# Make a copy of `selected_data` 
extracted_data = selected_data.copy()

# Add a new column named `total_points`; 
# Calculate total points earned by multiplying the number of games played by the average number of points earned per game
extracted_data["total_points"] = extracted_data["gp"] * extracted_data["pts"]

# Add a new column named `efficiency`. Calculate efficiency by dividing the total points earned by the total number 
# of minutes played, which yields points per minute. (Note that `min` represents avg. minutes per game.)
extracted_data["efficiency"] = extracted_data["total_points"] / (extracted_data["min"] * extracted_data["gp"])

# Display the first few rows of `extracted_data` to confirm that the new columns were added.
extracted_data.head()
```
